```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# Define node labels
nodes = [
    "HTTP Request", 
    "API Endpoint", 
    "session_data", 
    "Router", 
    "Pipes (global, route-specific)",
    "MasterAgent Route", 
    "MasterAgent", 
    "HTTP Response to User"
]

# Add nodes to the graph
for node in nodes:
    G.add_node(node)

# Define edges based on the ASCII flow
edges = [
    ("HTTP Request", "API Endpoint"),
    ("API Endpoint", "session_data"),
    ("session_data", "Router"),
    ("Router", "Pipes (global, route-specific)"),
    ("Pipes (global, route-specific)", "MasterAgent Route"),
    ("MasterAgent Route", "MasterAgent"),
    ("MasterAgent", "Router"),  # If destination = another_route
    ("MasterAgent", "HTTP Response to User")  # If destination = 'user'
]

# Add edges to the graph
G.add_edges_from(edges)

# Create layout and draw
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True, node_size=3000, node_color="lightgray", arrowstyle="-|>", arrowsize=12)
plt.title("OLAF Data Flow Diagram")
plt.show()
```
